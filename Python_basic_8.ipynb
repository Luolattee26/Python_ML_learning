```python
# 9.1
# 此处的yield起到了一个“中继器/暂停器”的作用
# 对于函数中的return来说，类似于循环中的break，当计算机处理到return后，一定会返回一个值（空值或者设定好的值）
# 此时整个函数function会被清空，从内存中释放出来，变量之类的也会被删除
# 而yield则同样具有返回的功能，返回一个值，但是与return不同的是，yield返回值后会记住当前位置，而下一次接到指令时，能从当前位置重新启动
# yield此种作用特性就可以让函数“中途暂停“，由于yield在暂停时也可以清空内存，所以可以利用yield来观察运行情况，并且决定是否继续
# 回到本题中，如果将yield改为return，则在输入第一句话后ALL_upper()便会停止，此时在print语句中for string in ALL_upper()则等价为
# 'Hello'（为例子）这么一个字符串，最终结果就是HELLO这几个字母单个按行输出
# 如果使用yield，则在输入一句话后，会返回当前的string值，并且开始下一轮的句子输入，最终形成一个含有多个字符串的list，实现多行输出
def ALL_upper():
    while True:
        string = input('请输入需要大写处理的句子：')
        if not string:
            return
        yield string
    
if __name__ == '__main__':
    print(*(string.upper() for string in ALL_upper()),sep = '\n')
```

```python
# 10.1
def string_input(string):
    string_split = string.split(' ')
    return string_split


def remove_same(splicted_string):
    temp_lst = []
    for i in splicted_string:
        i_index = splicted_string.index(i)
        for j in range(i_index+1,len(splicted_string)-1): # 这里上限之所以需要-1，是因为下限有可能会取到i_index=9的情况
            j_word = splicted_string[j]
            if i == j_word:
                splicted_string.pop(j)
    return splicted_string
        
    
if __name__ == '__main__':
    str1 = string_input(input('请输入需要处理的语句，使用空格分割：'))
    str1_rmoved = remove_same(str1)
    str1_rmoved.sort()
    print(str1_rmoved)
    
    
# 下面使用list.count这个方法进行解决，是一种很高效的方法    
word = input().split()
for i in word:
    if word.count(i) > 1:    # count function returns total repeatation of an element that is send as argument
        word.remove(i)     # removes exactly one element per call
word.sort()
print(" ".join(word))


# 还有一种方法，使用set的特性不存在重复元素
word = sorted(list(set(input().split())))              
print(" ".join(word))
```

```python
# 13.1
def digit_letter(str_input):
    count_D = 0
    count_L = 0
    for i in str_input:
        judge_D = i.isdigit()
        judge_L = i.isalpha()
        if judge_D == True:
            count_D += 1
        elif judge_L == True:
            count_L += 1
        else:
            pass
    print('letter',count_L)
    print('digit',count_D)

    
if __name__ == "__main__":
    digit_letter(input('输入一个句子：'))
```

```python
# 14.1
def Upper_low_alpha(str_input):
    count_U = 0
    count_L = 0
    for i in str_input:
        judge = ord(i)
        if judge >= 65 and judge <= 90:
            count_U += 1
        elif judge >= 97 and judge <= 122:
            count_L += 1
        else:
            pass
    print('UPPER',count_U)
    print('LOWER',count_L)

    
if __name__ == "__main__":
    Upper_low_alpha(input('输入一个句子：'))
    

# 还有一种比较高级的办法
word = input()
upper = sum(1 for i in word if i.isupper()) # sum function cumulatively sum up 1's if the condition is True
lower = sum(1 for i in word if i.islower()) # 感觉也是类似生成器的语法，利用sum做不断循环

print("UPPER CASE {0}\nLOWER CASE {1}".format(upper,lower)) # 用了占位符
```

```python
# 16.1
def num_input(string):
    string_split = string.split(',')
    return string_split


def judge(num_str):
    square_lst = []
    for i in num_str:
        if int(i)%2 != 0:
            square_lst.append(i)
    return square_lst


def square(lst):
    out_lst = []
    for j in lst:
        out_lst.append(int(j)**2)
    print(out_lst)
    
    
if __name__ == '__main__':
    num_str = num_input(input('输入一串以逗号分隔的数字：'))
    square_lst = judge(num_str)
    square(square_lst)
```

```python
# 16.2
# 这里使用了filter函数，filter函数可以进行筛选
# 第一个参数为筛选标准函数，第二个参数为可迭代对象，如字符串、列表等
seq = input().split(',') 
lst = [int(i) for i in seq]
def flt(i):                                           # Define a filter function
    return i % 2 != 0
result_l = [str(i * i) for i in filter(flt,lst)]      
print(",".join(result_l))
```

```python
# 17.1
class calc_money:
    
    def log_input(self):
        log_list = []
        while True:
            log_str = input('请输入交易日志：')
            if len(log_str) <= 0:
                break
            else:
                log_list.append(log_str)
        return log_list

            
    def judge_DW(self,log_lst):
        Standardized_lst = []
        for i in range(len(log_lst)):
            log1 = log_lst[i]
            if log1[0] == 'D' or log1[0] == 'd':
                Standardized_lst.append(int(log1[2:len(log1)]))
            elif log1[0] == 'W' or log1[0] == 'w':
                Standardized_lst.append(-int(log1[2:len(log1)]))
            else:
                print('日志格式错误')
                break
        return Standardized_lst
                
                
    def final_calc(self,stand_data):
        result = 0
        for i in stand_data:
            result += i
        print(result)
        

if __name__ == '__main__':
    calc_tools = calc_money()
    lst1 = calc_tools.log_input()
    lst2 = calc_tools.judge_DW(lst1)
    calc_tools.final_calc(lst2)
```

```python
# 18.2
# 这里用了re这个模块
import  re

s = input().split(',')
lst = []

for i in s:
    cnt = 0
    cnt+=(6<=len(i) and len(i)<=12)
    cnt+=bool(re.search("[a-z]",i))      # here re module includes a function re.search() which returns the object information
    cnt+=bool(re.search("[A-Z]",i))      # of where the pattern string i is matched with any of the [a-z]/[A-z]/[0=9]/[@#$] characters
    cnt+=bool(re.search("[0-9]",i))      # if not a single match found then returns NONE which converts to False in boolean
    cnt+=bool(re.search("[@#$]",i))      # expression otherwise True if found any.
    if cnt == 5:
        lst.append(i)

print(",".join(lst))
```
